```python
import pandas as pd
import seaborn as sns

from fwi_predict.geo.ee import SENTINEL2_SCL_MAP
```

```python
sentinel_features = pd.read_csv('../data/raw/sentinel2_features.csv').drop(columns=['system:index', '.geo'])
```

```python
sentinel_features['SCL'] = sentinel_features['SCL'].map(SENTINEL2_SCL_MAP)
```

Check scene classification

```python
ax = sns.histplot(sentinel_features, x='SCL')
ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
```
